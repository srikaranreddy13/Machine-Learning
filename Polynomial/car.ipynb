```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv("CarPrice_Assignment.csv")

X = data[['enginesize']]
y = data['price']
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

```python
X_range = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)

y_linear = linear_model.predict(X_range)
y_poly = poly_model.predict(poly.transform(X_range))
```

```python
plt.figure(figsize=(10, 6))

# Actual data
plt.scatter(X, y, alpha=0.6, label="Actual Data Points")

# Linear regression line
plt.plot(X_range, y_linear, color='red', linewidth=2,
         label="Linear Regression Fit")

# Polynomial regression curve
plt.plot(X_range, y_poly, color='green', linewidth=2,
         label="Polynomial Regression (Degree 4) Fit")

plt.xlabel("Engine Size")
plt.ylabel("Car Price")
plt.title("Engine Size vs. Car Price with Model Fits")
plt.legend()
plt.grid(True)

plt.show()
```
